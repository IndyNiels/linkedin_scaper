```python
## open response.json from the root directory 

import json

with open('response.json') as f:
    data = json.load(f)
```

```python
included = data['included']
```

```python
def print_keys(object):
    for key, value in object.items(): 
        print(key)
        print(value)
        ## if statement to check if item is a dictionary
        if isinstance(value, dict):
            print_keys(value)
        if isinstance(value, list):
            for item in value:
                print_keys(item)

print_keys(included[53])
  
## remove duplicates from the list of keys
```

```python
def find_keys(obj, target_value):
    for key, value in obj.items():
        if value == target_value:
            return True
        if isinstance(value, dict):
            if find_keys(value, target_value):
                return True
        if isinstance(value, list):
            for item in value:
                if isinstance(item, dict):
                    if find_keys(item, target_value):
                        return True
    return False

target_value = 'Promoted'

# Iterate over the list and store indices of matching objects
matching_indices = []
for index, obj in enumerate(included):
    if find_keys(obj, target_value):
        matching_indices.append(index)

# Print the result
if matching_indices:
    print("Indices of matching objects:", matching_indices)
else:
    print("No objects with matching key found.")
```
